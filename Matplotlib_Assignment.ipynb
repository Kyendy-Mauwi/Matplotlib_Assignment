```python
#imporing libs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#loading the dataset
data = 'jumta.csv'
df = pd.read_csv(data)
df.head(5)
```

```python
#line graph showing monthly rev 
plt.plot(df['Month'], df['Revenue'], label='Revenue')
plt.title('Monthly Revenue 2020-2021')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
#line graph showing monthly rev with grid
plt.plot(df['Month'], df['Revenue'], marker='*', linestyle='--', label='Revenue')
plt.title('Monthly Revenue 2020-2021')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.xticks(rotation=90)
plt.legend()
plt.grid()
plt.show()
```

```python
#bar plot showing total number of orders per month
plt.bar(df['Month'], df['Orders'], label='Revenue')
plt.title('Monthly Orders 2020-2021')
plt.xlabel('Month')
plt.ylabel('Order number')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

```python
#Scatter plot, r/ship btwn no. of orders and avg order value
plt.scatter(df['Orders'], df['Average_Order_Value'], label='Order Value')
plt.title('Num of Orders vs Avg Order value')
plt.xlabel('Num of Orders')
plt.ylabel('Avg Order Value')
plt.grid()
plt.legend()
plt.show()
```

```python
#subplot containing all visualizations
fig, (one, two, three) = plt.subplots(nrows=3, ncols=1, figsize=(12, 18))

one.plot(df['Month'], df['Revenue'], label='Revenue')
one.set_title('Monthly Revenue')
one.set_xlabel('Month')
one.set_ylabel('Revenue')
one.tick_params(axis='x', rotation=90)
one.grid()
one.legend()

two.bar(df['Month'], df['Orders'], label='Orders')
two.set_title=('Monthly Orders')
two.set_xlabel('Month')
two.set_ylabel('Num of Orders')
two.tick_params(axis='x', rotation=90)
two.grid()
two.legend()

three.scatter(df['Orders'], df['Average_Order_Value'], label='Order Value')
three.set_title('Num of Orders vs Avg Order value')
three.set_xlabel('Num of Orders')
three.set_ylabel('Avg Order Value')
three.grid(axis='y')
three.legend()

plt.tight_layout()
plt.savefig('Sales Summary', dpi=300)
plt.show()
```
